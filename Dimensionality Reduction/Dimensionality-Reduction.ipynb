```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs, make_circles
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA, KernelPCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.manifold import Isomap
from sklearn.manifold import TSNE
```

```python
# Generate and plot the original data
attributes, _ = make_blobs(centers = 1)
#standart_scaler = StandardScaler()
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(attributes)
#scaled_data_standard = standart_scaler.fit_transform(attributes)
plt.scatter(scaled_data[:, 0], scaled_data[:, 1])
plt.show()
```

```python
pca = PCA().fit(scaled_data)
transformed_data = pca.transform(scaled_data)
plt.scatter(transformed_data[:, 0], transformed_data[:, 1])
plt.show()
pca.explained_variance_ratio_
```

# Kernel PCA

```python
kernel_pca = KernelPCA(kernel = "rbf", gamma = 5).fit(attributes)
kernel_transformed_data = pca.transform(attributes)
plt.scatter(kernel_transformed_data[:, 0], kernel_transformed_data[:, 1])
plt.show()
```

# Linear Discriminant Analysis (LinDA)

```python
lda = LinearDiscriminantAnalysis(n_components = 2)
# transformed_attributes = lda.fit(attributes).transform(attributes)
```

# Isometric Mapping

```python
isomap = Isomap(n_neighbors = 5, n_components = 2)
transformed_data = isomap.fit_transform(attributes)
plt.scatter(transformed_data[:, 0], transformed_data[:, 1])
plt.show()
```

# t-SNE

```python
tsne = TSNE(n_components = 2, init = "pca")
transformed_data = tsne.fit_transform(attributes)
plt.scatter(transformed_data[:, 0], transformed_data[:, 1])
plt.show()
```
